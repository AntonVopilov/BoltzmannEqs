```python
import pickle
import matplotlib.pyplot as plt
import os,glob
import numpy as np
import seaborn as sns
from scipy.special import kn,zetac
sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

```python
pts = []
for f in glob.glob('testSimple*dat'):
    lamb = eval(f.split('_')[1].split('.dat')[0])
    with open(f,'r') as ff:
        omega = eval([l for l in ff.readlines() if "Omega" in l][0].split('=')[-1])
    pts.append([lamb,omega])
pts = np.array(sorted(pts))
```

```python
plt.plot(pts[:,0],pts[:,1],'r-',linewidth=3)
plt.plot(pts[:,0],[0.12]*len(pts),'--',linewidth=3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$\Omega_{DM} h^2$')
plt.savefig("FO_FI_transition.png")
plt.show()
```
