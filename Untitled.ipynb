```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.genfromtxt('test.dat',names=True,skip_header=10)
```

```python
print(d.dtype.names)
```

```python
plt.plot(d['R'],d['n_DM'],'b--',label='DM')
plt.plot(d['R'],d['n_Mediator'],'r--',label='Mediator')
plt.xscale('log')
plt.yscale('log')
plt.legend(loc='lower right',fontsize=15)
plt.show()
```

```python
plt.plot(d['R'],d['rho_DM'],'b--',label='DM')
plt.plot(d['R'],d['rho_Mediator'],'r--',label='Mediator')
plt.plot(d['R'],d['T'],label='T')
plt.xscale('log')
plt.yscale('log')
plt.legend(loc='lower right',fontsize=15)
plt.show()
```
