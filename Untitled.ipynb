```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.loadtxt('default.txt')
n = np.loadtxt('new.txt')
```

```python
plt.plot(d[:,0],d[:,3],'b--')
plt.plot(n[:,0],n[:,5],'r-.')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
plt.plot(d[:,0],d[:,4],'b--')
plt.plot(n[:,0],n[:,4],'r-.')
plt.xscale('log')
plt.yscale('log')
plt.xlim(3000.,4000.)
plt.ylim(1e-10,10.)
plt.show()
```

```python
plt.plot(d[:,1],d[:,5],'b--')
plt.plot(n[:,2],n[:,6],'r-.')
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.01,100.)
plt.show()
```
